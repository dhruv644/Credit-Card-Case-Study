### Cleaning Age columns

```python
credit_card['Age']=credit_card.Age.astype("int")
credit_card["Age"] = np.where(credit_card.Age< 18 , 18, credit_card.Age)
customer['Age'] = np.where(customer.Age< 18 , 18, customer.Age)
```

```python
cust_spend = pd.merge(left=spend,right=customer,on="Customer")
cust_spend.Amount = np.where(cust_spend.Amount>cust_spend.Limit,cust_spend["Limit"]*0.5,cust_spend.Amount)
spend.loc[cust_spend.Amount>cust_spend.Limit,"Amount"] = cust_spend.Amount
```
